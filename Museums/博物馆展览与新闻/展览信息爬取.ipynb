```python
import datetime,time,random,os
import requests,re
from bs4 import BeautifulSoup
import pandas as pd
import json
from tqdm.notebook import tqdm
```

爬取新闻

```python
res=[]
for i in tqdm(range(710)):
    url='https://www.chinamuseum.org.cn/newList.html?id=13&pageIndex='+str(i+1)
    r=requests.get(url)
    result=BeautifulSoup(r.content,'lxml') 
    a=result.find('ul',{'class':'titlelist'}).find_all('li')
    for i in a:
        try:
            title=i.find('a').get_text()
            url=i.find('a').get('href')
            time=i.find('span').get_text()
            res.append([title,url,time])
        except:
            continue
df=pd.DataFrame(data=res,columns=['title','url','time'])
df
```

```python
content=[]
for i in tqdm(range(len(df))):
    try:
        r=requests.get('http:'+df['url'].iloc[i])
        result=BeautifulSoup(r.content,'lxml') 
        a=result.find('div',{'class':'post-desc'})
        content.append(a.get_text())
    except:
        content.append('Error')
```

```python
df['content']=content
df.to_excel('博物馆近年新闻.xlsx',index=False)
```

爬取展览信息

```python
res=[]
for i in tqdm(range(168)):
    url='https://www.chinamuseum.org.cn/cma/photoList.html?id=14&pageIndex='+str(i+1)
    r=requests.get(url)
    result=BeautifulSoup(r.content,'lxml') 
    a=result.find('div',{'class':'photoList'}).find_all('li')
    for i in a:
        try:
            title=i.find('div').find('h4').get_text()
            url=i.find('div').find('a').get('href')
            des=i.find('div').find('p').get_text()
            detail=i.find('div').find('div').get_text()
            res.append([title,url,des,detail])
        except:
            continue
df=pd.DataFrame(data=res,columns=['title','url','description','detail'])
df
```

```python
content=[]
for i in tqdm(range(len(df))):
    try:
        r=requests.get('http:'+df['url'].iloc[i])
        result=BeautifulSoup(r.content,'lxml') 
        a=result.find('div',{'class':'post-desc'})
        content.append(a.get_text())
    except:
        content.append('Error')
```

```python
df['content']=content
df.to_excel('博物馆近年展览.xlsx',index=False)
```
