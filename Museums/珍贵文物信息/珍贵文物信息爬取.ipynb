```python
import datetime,time,random,os
import requests,re
from bs4 import BeautifulSoup
import pandas as pd
import json
 
startPage="http://gl.ncha.gov.cn:9997/portal/collection/cr/find"

 
myHeader={
'Accept': 'application/json, text/plain, */*',
'Accept-Encoding': 'gzip, deflate',
'Accept-Language': 'zh-CN,zh;q=0.9,en;q=0.8',
'Connection': 'keep-alive',
'Content-Length': '89',
'Content-Type': 'application/json;charset=UTF-8',
'Host': 'gl.ncha.gov.cn:9997',
'Origin': 'http://gl.ncha.gov.cn',
'Referer': 'http://gl.ncha.gov.cn/',
'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36'
}
```

```python
data1=json.dumps({
  "sort": {
    "collectionName": "1"
  },
  "pageNation": {
    "currentPage": 1,
    "pageSize": 2360000
  },
  "keyWord": ""
})
```

```python
r=requests.post(startPage,data = data1,headers=myHeader)
```

```python
result=r.json()
```

```python
fileObject = open('/content/drive/MyDrive/museum/zhenguiwenwu.json', 'w',encoding='utf-8') 
fileObject.write(json.dumps(result,ensure_ascii=False))  
fileObject.close()
```
